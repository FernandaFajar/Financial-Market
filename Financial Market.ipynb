## Daily Return

```python
# Getting daily return
# Saham OCBC NISP
nisp = qs.utils.download_returns('NISP.JK')
nisp = nisp.loc['2010-01-01':'2023-10-01']

# Saham Berkah Prima Perkasa (Blueprint)
blue = qs.utils.download_returns('BLUE.JK')
blue = blue.loc['2010-01-01':'2023-10-01']

# Saham Gojek Tokopedia (GOTO)
goto = qs.utils.download_returns('GOTO.JK')
goto = goto.loc['2010-01-01':'2023-10-01']



# Make the datetime index timezone-aware
nisp.index = pd.to_datetime(nisp.index).tz_localize(None)
blue.index = pd.to_datetime(blue.index).tz_localize(None)
goto.index = pd.to_datetime(goto.index).tz_localize(None)

# Define a benchmark (e.g., S&P 500, symbol: 'SPY')
benchmark = qs.utils.download_returns('IDX')
benchmark = benchmark.loc['2010-01-01':'2023-10-01']

# Make the datetime index timezone-aware
benchmark.index = pd.to_datetime(benchmark.index).tz_localize(None)

# Plotting daily returns with benchmark
print('\nNISP Daily Returns Plot:\n')
qs.plots.daily_returns(nisp, benchmark=benchmark)

print('\nBLUE Daily Returns Plot:\n')
qs.plots.daily_returns(blue, benchmark=benchmark)

print('\nGOTO Daily Returns Plot:\n')
qs.plots.daily_returns(goto, benchmark=benchmark)
```

## Optimizing Portfolio

```python
#Getting dataframes info for stocks using yfinance
nisp_df = yf.download('NISP.JK', start='2010-01-01', end='2023-10-01')['Adj Close']
blue_df = yf.download('BLUE.JK', start='2010-01-01', end='2023-10-01')['Adj Close']
goto_df = yf.download('GOTO.JK', start='2010-01-01', end='2023-10-01')['Adj Close']
```

```python
# Merging and creating an 'Adj Close' DataFrame
df = pd.concat([nisp_df, blue_df, goto_df], axis=1, join='outer')
df.columns = ['nisp', 'blue', 'goto']

# Display the resulting DataFrame
print(df)
```

```python
# Importing libraries for portfolio optimizatin
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns
```

# Markowitz Mean-Variance Optimization Model
